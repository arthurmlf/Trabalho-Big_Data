```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados_excel = pd.read_excel('/content/drive/MyDrive/Trabalho de Big Data/Relatorio de vendas.xlsx')
print(dados_excel)
```

```python
df = pd.read_excel('/content/drive/MyDrive/Trabalho de Big Data/Relatorio de vendas.xlsx')

ordem_dias_semana = ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
cores = ['red', 'green', 'blue', 'purple', 'orange', 'brown']

df['Dia_da_sem'] = pd.Categorical(df['Dia_da_sem'], categories=ordem_dias_semana, ordered=True)

vendas_por_dia = df.groupby('Dia_da_sem')['Valor'].sum()

plt.bar(vendas_por_dia.index, vendas_por_dia.values, width = 0.5, color = cores)
plt.xlabel('Dia da Semana')
plt.ylabel('Total de Vendas')
plt.title('Total de Vendas por Dia da Semana')
plt.savefig('total_vendas_dias.png', dpi=400)
plt.show()
```

```python
vendas_por_mes = df.groupby(df['Mês'])['Valor'].sum()

cores = ['blue', 'green', 'red', 'purple', 'orange', 'brown',
         'gray', 'pink', 'cyan', 'magenta', 'yellow', 'olive']

vendas_por_mes.plot(kind='bar', color = cores)
plt.xlabel('Mês')
plt.ylabel('Total de Vendas')
plt.title('Total de Vendas por Mês')
plt.xticks(rotation=45)
plt.savefig('total_vendas_mes.png', dpi=400)
plt.show()
```

```python
ticket_medio_por_mes = df.groupby('Mês')['Valor'].mean()

ticket_medio_por_mes.plot(kind='bar', color = 'purple')

plt.title('Ticket Médio de Vendas por Mês')
plt.xlabel('Mês')
plt.ylabel('Ticket Médio')
plt.savefig('ticket_med_mes.png', dpi=400)
plt.show()
```

```python
percentual_ocorrencia = df['Tipo_venda'].value_counts(normalize=True) * 100

cores = ['purple', 'magenta']

percentual_ocorrencia.plot(kind='pie', autopct='%1.1f%%', colors=['magenta', 'pink'])
plt.title('Percentual de Vendas Online e Loja Física')
plt.ylabel('')
plt.savefig('percentual_de_vendas_tipo.png', dpi=400)
plt.show()
```
